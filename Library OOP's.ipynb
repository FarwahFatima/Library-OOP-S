```python
class Farwah_Library:
    dct = {}                       # dictinary for adding and storing books
    def __init__(self,name,list):
        self.name = name
        self.list = list
        
    @staticmethod                
    def continuee():
        print('\n')
        print("WELCOME TO OUR FARWAH'S LIBRARY")
        print('What operation you want to  perform? ')
        print('1. Display Books')
        print('2. Lend a Book')
        print('3. Add a Boook')
        print('4. Return a  book')
        
    def Display(self):             # To display Books
        return f'The available books are \n{self.list}'
    
    def Lend(self):                # Lend Book with name
        book_name = input('Enter a book you want to lend ?\n')
        user_name = input('Whats your name ?')
        if book_name in self.list:
            if book_name not in self.dct.keys():
                self.dct.update({book_name:user_name})
                return f"Done! you can now lend {book_name}"
            else:
                return f'the book is owned by {self.dct[book_name]}'
        else:
            return f'Sorry the book you entered is not in our Library'
    
    def Add(self):                # Add Book with  their name
        name_book = input('Enter name of the book you want to add: ')
        self.list.append(name_book)
        return f'{name_book} has been added'
    
    def Return(self):             # Return a Book with their name
        name_book = input('Enter book you want to return: ')
        name_user = input('\nEnter your  name: ')
        if name_book in self.dct.keys() and name_user in self.dct.values():
            self.dct.pop(name_book)
            return f'Successfully  returned the book'
        else:
            return f'Sorry {name_user} you havee not lend {name_book}'
        
books = Farwah_Library('Farwah',['The Power of Focus','The Secret if the Grown-up Brain','Ego is the Enemy','Ideology of the Future'])        

while True:
    books.continuee()
    n = input("Enter a number: ")
    if n == '1':
        print(books.Display())
    elif n == '2':
        print(books.Lend())
    elif n == '3':
        print(books.Add())
    elif n == '4':
        print(books.Return())
    else :
        print('The number is not allowed')
        
    a = input("Do you want to continue? \nEnter y for yes or n for no \n")    
    if a == 'y':
        continue
    elif a == 'n':
        break
    else: 
        print("Please enter only y or n \n")
```
